## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

```python
# Convert price_date to datetime if not already
price_df['price_date'] = pd.to_datetime(price_df['price_date'])

# Merge dataframes
merged_df = pd.merge(client_df, price_df, on='id', how='left')


```

```python
# Assuming price_date is already in datetime format after the merge
average_prices = merged_df.groupby('price_date').agg({
    'price_off_peak_var': 'mean',
    'price_peak_var': 'mean',
    'price_mid_peak_var': 'mean'
}).reset_index()

plt.figure(figsize=(12, 6))
plt.plot(average_prices['price_date'], average_prices['price_off_peak_var'], label='Off Peak Variable Price')
plt.plot(average_prices['price_date'], average_prices['price_peak_var'], label='Peak Variable Price')
plt.plot(average_prices['price_date'], average_prices['price_mid_peak_var'], label='Mid Peak Variable Price')
plt.title('Average Variable Energy Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
client_df.replace([np.inf, -np.inf], np.nan, inplace=True)
plt.figure(figsize=(10, 6))
sns.kdeplot(client_df[client_df['churn'] == 0]['num_years_antig'], label='Retained', fill=True)
sns.kdeplot(client_df[client_df['churn'] == 1]['num_years_antig'], label='Churned', fill=True)
plt.title('Customer Tenure Distribution by Churn')
plt.xlabel('Tenure in Years')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(client_df['cons_gas_12m'], kde=True, bins=50, color='green')
plt.title('Histogram of Gas Consumption in the Last 12 Months')
plt.xlabel('Gas Consumption')
plt.ylabel('Frequency')
plt.show()
```

```python
gas_clients = client_df['has_gas'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(gas_clients, labels=gas_clients.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Gas Clients')
plt.show()
```

```python
price_df['price_date'] = pd.to_datetime(price_df['price_date'])
average_price_trend = price_df.groupby('price_date')['price_off_peak_var'].mean()

plt.figure(figsize=(12, 6))
average_price_trend.plot()
plt.title('Average Off-Peak Price Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Average Off-Peak Price')
plt.show()
```

```python
# Calculating monthly averages for consumption and prices
monthly_avg = merged_df.groupby(merged_df['price_date'].dt.to_period('M')).agg({
    'cons_last_month': 'mean',
    'price_off_peak_var': 'mean',
    'price_peak_var': 'mean'
}).reset_index()

monthly_avg['price_date'] = monthly_avg['price_date'].dt.to_timestamp()

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Avg Consumption (Last Month)', color=color)
ax1.plot(monthly_avg['price_date'], monthly_avg['cons_last_month'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Price', color=color)  # we already handled the x-label with ax1
ax2.plot(monthly_avg['price_date'], monthly_avg['price_off_peak_var'], label='Off Peak Variable Price', color='blue')
ax2.plot(monthly_avg['price_date'], monthly_avg['price_peak_var'], label='Peak Variable Price', color='green')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Average Monthly Consumption vs. Pricing Over Time')
plt.legend(loc='upper left')
plt.show()
```
